```python
def wb_distance_list(x):
    try:
        return ([round(distance.distance((x.latitude, x.longitude), 
                                         (coord[1], coord[0])).km, 5) 
                 for coord in x.nhd_vars_wb_centroid[1]])
    except:
        return []
    
def fl_distance_list(x):
    try:
        return ([round(distance.distance((x.latitude, x.longitude), 
                                         (coord[1], coord[0])).km, 5) 
                 for coord in x.nhd_vars_fl_centroid[1]])
    except:
        return []
    
# df_centroid["wb_centroid_dist_list"] = (df_centroid.apply(lambda x: 
#                                                           [round(distance.distance((x.latitude, x.longitude), 
#                                                                                    (coord[1], coord[0])).km, 5) 
#                                                            for coord in x.nhd_vars_wb_centroid[1]], 
#                                                           axis=1))

df_centroid["wb_centroid_dist_list"] = (df_centroid.apply(lambda x: wb_distance_list(x), axis=1))
df_centroid["fl_centroid_dist_list"] = (df_centroid.apply(lambda x: fl_distance_list(x), axis=1))
```

```python
def wb_min_index(x):
    try:
        return int(x.wb_centroid_dist_list.index(min(x.wb_centroid_dist_list)))
    except:
        return np.nan
    
def fl_min_index(x):
    try:
        return int(x.fl_centroid_dist_list.index(min(x.fl_centroid_dist_list)))
    except:
        return np.nan

df_centroid["wb_min_index"] = df_centroid.apply(lambda x: wb_min_index(x), axis=1)
df_centroid["fl_min_index"] = df_centroid.apply(lambda x: fl_min_index(x), axis=1)


```

```python
def closest_wb_distance(x):
    try:
        return x.wb_centroid_dist_list[int(x.wb_min_index)]
    except:
        return np.nan

def closest_fl_distance(x):
    try:
        return x.fl_centroid_dist_list[int(x.fl_min_index)]
    except:
        return np.nan
    
df_centroid["closest_wb_distance_m"] = df_centroid.apply(lambda x: closest_wb_distance(x), axis=1)
df_centroid["closest_fl_distance_m"] = df_centroid.apply(lambda x: closest_fl_distance(x), axis=1)
```

```python
def closest_wb_area(x):
    try:
        return x.nhd_vars_wb_centroid[3][int(x.wb_min_index)]
    except:
        return np.nan


# fl coord need to be extracted from nhd_stats
# so need one additional function
    
def extract_feature(comid, feature):
    """
    Extract features present in nhd_stats for corresponding comid
    """
    if comid == None:
        return np.nan # if no comid's in GEE
    extracted_feature = nhd_stats[nhd_stats["comid"] == comid][str(feature)]
    try:
        extracted_feature = np.array(extracted_feature).item() 
    except Exception as e:
        return np.nan # if comid in GEE but not in nhd database
    return extracted_feature
    
    
def closest_fl_area(x):
    try:
        return extract_feature(x.nhd_vars_fl_centroid[0][int(x.fl_min_index)], "areasqkm")
    except:
        return np.nan
    
df_centroid["closest_wb_area_sqkm"] = df_centroid.apply(lambda x: closest_wb_area(x), axis=1)
df_centroid["closest_fl_area_sqkm"] = df_centroid.apply(lambda x: closest_fl_area(x), axis=1)
```

```python
def closest_wb_centroid(x):
    try:
        return x.nhd_vars_wb_centroid[1][int(x.wb_min_index)]
    except:
        return np.nan

def closest_fl_centroid(x):
    try:
        return x.nhd_vars_fl_centroid[1][int(x.fl_min_index)]
    except:
        return np.nan
    
df_centroid["closest_wb_centroid_coord"] = df_centroid.apply(lambda x: closest_wb_centroid(x), axis=1)
df_centroid["closest_fl_centroid_coord"] = df_centroid.apply(lambda x: closest_fl_centroid(x), axis=1)
```

```python
def square(lat=45.475649, lon=-69.471018, size=100):
  crs_proj = "EPSG:4326"  
  return ee.Geometry.Point([lon, lat], proj=crs_proj).buffer(size).bounds()


srtm = ee.Image('USGS/SRTMGL1_003')
# slope = ee.Terrain.slope(srtm)
# jrc = ee.Image("JRC/GSW1_2/GlobalSurfaceWater").select("seasonality", "recurrence")

def get_wb_elevation(x, patch_size=10):
    image = ee.Image('USGS/SRTMGL1_003')
    patch_size = 10
    try:
        lat = x.closest_wb_centroid_coord[1]
        lon = x.closest_wb_centroid_coord[0]
        mean = image.reduceRegion(
                reducer = ee.Reducer.mean(),
                geometry = square(lat, lon, patch_size),
                scale = 30,
                maxPixels = 1e9
            ).getInfo()

        return mean.get('elevation')

    except Exception as e:
#         print(e)
        return np.nan

    
def get_fl_elevation(x, patch_size=10):
    image = ee.Image('USGS/SRTMGL1_003')
    patch_size = 10
    try:
        lat = x.closest_fl_centroid_coord[1]
        lon = x.closest_fl_centroid_coord[0]
        mean = image.reduceRegion(
                reducer = ee.Reducer.mean(),
                geometry = square(lat, lon, patch_size),
                scale = 30,
                maxPixels = 1e9
            ).getInfo()

        return mean.get('elevation')

    except Exception as e:
#         print(e)
        return np.nan
    
```

```python
# pass in batches of 500
# MADHUKAR: 0 - 5000 
# SHOBHA: 5000 - 10000
# RADHIKA: 10000 - 15000

patch_size = 10
batch_size = 500
MY_NAME = "MADHUKAR"
START = 0 + 5000 * (MY_NAME == "SHOBHA") + 10000 * (MY_NAME == "RADHIKA")

for batch in range(10):
  print("batch {} of 10 started".format(batch + 1))
  batch_df = df_centroid[START + batch_size * batch : START + batch_size * (batch + 1)].copy()
  batch_df["closest_wb_elevation"] = batch_df.apply(lambda x: get_wb_elevation(x, patch_size), axis=1)
  batch_df["closest_fl_elevation"] = batch_df.apply(lambda x: get_fl_elevation(x, patch_size), axis=1)
  print("batch {} of 10 done".format(batch + 1))
  i = (START + batch_size * batch) + 1
  (pickle.dump(batch_df, open("nearest_wb_fl_elevation/nearest_wb_fl_elevation"
                           + "_part" + str(i),"wb"), 
               protocol=3))          
    
```
