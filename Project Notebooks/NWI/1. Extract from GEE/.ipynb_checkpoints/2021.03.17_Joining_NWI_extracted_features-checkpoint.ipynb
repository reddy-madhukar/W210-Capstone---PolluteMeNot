```python
import pickle
import pandas as pd
```

```python
df_200m = pd.read_pickle("nwi_extracted_features_200X200m")
df_2500m = pd.read_pickle("nwi_extracted_features_2500X2500m")
df = pd.read_csv("combined_regular_clean_with_ssurgo_variables.csv")
```

```python
col_200m_delta = ["Index"] + list(set(df_200m.columns) - set(df.columns)) 
col_2500m_delta = ["Index"] + list(set(df_2500m.columns) - set(df.columns)) 

df_200m_delta = df_200m[col_200m_delta]
df_2500m_delta = df_2500m[col_2500m_delta]
```

```python
df1 = df
df2 = df_200m_delta
df3 = df_2500m

df_merged = pd.merge(df1, df2, on="Index")
df_merged = pd.merge(df_merged, df3, on="Index")
# df_merged = pd.merge(df_constant, df_merged, on="Index")
set(df_merged.columns)
```

```python
df_merged.shape
```

```python
import pickle

# use protocol 3 for backwards compatibility with Python 3.6 on AWS
pickle.dump(df_merged, open("2021.03.09_nwi_extracted_features","wb"), protocol=3)
df_merged.to_csv("2021.03.09_nwi_extracted_features.csv")
```
